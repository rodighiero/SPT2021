
# Masked wordcloud

Using a mask you can generate wordclouds in arbitrary shapes.


```python
from os import path
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import os
import numpy as np

from wordcloud import WordCloud, STOPWORDS
from PIL import Image, ImageDraw

text = 'Lorem ipsum dolor sit amet, consectetur adipiscing elit, sed do eiusmod tempor incididunt ut labore et dolore magna aliqua. Ut enim ad minim veniam, quis nostrud exercitation ullamco laboris nisi ut aliquip ex ea commodo consequat. Duis aute irure dolor in reprehenderit in voluptate velit esse cillum dolore eu fugiat nulla pariatur. Excepteur sint occaecat cupidatat non proident, sunt in culpa qui officia deserunt mollit anim id est laborum.'


# Create mask

width = 400; height = 400
points = [(0, 300), (300, 300), (300,0)]
img = Image.new(mode = "RGBA", size = (width, height), color = (255, 255, 255))
draw = ImageDraw.Draw(img)
draw.polygon(points, fill=(0,0,0))
mask =np.array(img)


# Create wordcloud

stopwords = set(STOPWORDS)
stopwords.add("said")
wc = WordCloud(mode = "RGBA", background_color=None, max_words=2000, mask=mask, stopwords=stopwords, contour_width=0, scale=2)
wc.generate(text) # generate word cloud
wc.to_file(path.join("mask.png")) # store to file


# Show result

plt.imshow(wc, interpolation="bilinear")
plt.margins(0)
plt.axis('off')
plt.show()
```
